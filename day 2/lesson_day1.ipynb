```python
def even_or_odd(a):
    """ this function evaluate whether the number is even or odd"""
    if a % 2 == 0:
        print(f'{a} is even')
    else:
        print(f'{a} is odd')
    
```

```python
even_or_odd(44)
```

```python
def even_or_odd2(a):
    """ this function evaluate whether the number is even or odd"""
    if a % 2 == 0:
        print(f'{a} is even')
    
    print(f'{a} is odd')
```

```python
even_or_odd2(59)
```

```python
def even_or_odd3(a):
    """ this function evaluate whether the number is even or odd
    it returns true if the number is even and false when the number is false """
    return a % 2 == 0
```

```python
even_or_odd3(50)
```

```python
df = pd.read_csv('Online Retail.csv', sep =';', encoding = 'ISO-8859-1')
df.head()
```

```python
df = df[df['Quantity'] > 0]
df
```

```python
df = df[df['UnitPrice'] > 0]
df
```

```python
df
```

```python
df['UnitPrice'] = (
    df['UnitPrice']
    .str.replace(',', '.', regex=False)
    .astype(float)
)

df['Quantity'] = df['Quantity'].astype(int)

df['TotalValue'] = df['Quantity'] * df['UnitPrice']
```

```python
df['Quantity'].min()
```

```python
df['UnitPrice'].min()
```

```python
weekly_sales2 = df['TotalValue'].resample('W').sum()
```

```python
weekly_sales2
```

```python
weekly_sales2.shape
```

```python
!pip install statsmodels
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
weekly_sales2 = weekly_sales2[weekly_sales2 != 0]
```

```python
weekly_sales2
```

```python
weekly_sales2.shape
```

```python
weekly_sales2.plot(figsize=(10,4), title='Weekly Sales')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomp = seasonal_decompose(weekly_sales2, model='additive', period=26)  # 52 settimane = 1 anno
decomp.plot()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model = ARIMA(weekly_sales2, order=(1,1,1))
model_fit = model.fit()
print(model_fit.summary())
```

```python
forecast = model_fit.get_forecast(steps=12)
forecast_values = forecast.predicted_mean
forecast_ci = forecast.conf_int()

print(forecast_values)
```

```python
print(forecast_ci)
```

```python
forecast = list(forecast_values)
```

```python
plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:,0],   # lower
    forecast_ci.iloc[:,1],   # upper
    color='green', alpha=0.2, label='95% confidence interval'
)
plt.plot(forecast_values.index, forecast_values, label='Expected values', color='black', marker='o')
```

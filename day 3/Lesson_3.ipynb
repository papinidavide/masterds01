# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,3,3,4,5]
y = [9,3,5,1,10]

plt.plot(x,y)
plt.title('Graph')
plt.xlabel('Days')
plt.ylabel('Rain mm')
plt.show()
```

```python
categories = ['company a', 'company b', 'company c', 'company d']
values = [20,50,-11,40]

colors = ['green', 'blue', 'red', 'orange']

plt.barh(categories,values, color = colors, edgecolor = 'black')
plt.title('Bar chart')
plt.xlabel('North-west companies',)
plt.ylabel('Revenues')
plt.xticks(rotation = 45)
plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

categories = ['company a', 'company b', 'company c', 'company d']
values = [20,50,-11,40]

colors = ['green', 'blue', 'red', 'orange']

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = np.arange(len(categories))

ax.bar(x,values, zdir = 'y',color = colors)

# plt.title('Bar chart')
# plt.xlabel('North-west companies',)
# plt.ylabel('Revenues')
plt.xticks(x)
plt.show()
```
